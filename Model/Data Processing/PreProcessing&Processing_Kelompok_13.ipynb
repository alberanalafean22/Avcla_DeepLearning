### Cacomantis Merulinus

```python
pip install pydub
```

```python
import os
from google.colab import drive
from pydub import AudioSegment

# Mount Google Drive
drive.mount('/content/drive')

# Input and output folder paths
input_folder = "/content/drive/MyDrive/Filter Data/Cacomantis Merulinus"
output_folder = "/content/drive/MyDrive/Dataset Burung Resampling/Cacomantis Merulinus"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate through all files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.mp3'):  # Add other audio file extensions if needed
        input_file = os.path.join(input_folder, filename)

        try:
            # Load audio file using pydub
            audio = AudioSegment.from_file(input_file)

            # Convert to mono and 16kHz
            audio = audio.set_channels(1).set_frame_rate(16000)

            # Construct the output filename (replace original extension with .wav)
            name, ext = os.path.splitext(filename)
            output_file = os.path.join(output_folder, f"{name}.wav")

            # Export the audio file
            audio.export(output_file, format="wav")
            print(f"Processed: {filename} -> {output_file}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

## Segmentasi (pemotongan durasi hanya mengambil 5 detik)

###Cacomantis Merulinus

```python
pip install pydub
```

```python
import os
from pydub import AudioSegment
from pydub.utils import make_chunks

def split_audio(input_folder, output_folder, chunk_length_ms=5000):
    # Cek apakah folder output ada, jika tidak buat
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterasi semua file dalam folder input
    for filename in os.listdir(input_folder):
        file_path = os.path.join(input_folder, filename)

        # Periksa jika file adalah file audio
        if filename.endswith(".wav"):
            print(f"Memproses: {filename}")

            # Memuat file audio
            audio = AudioSegment.from_file(file_path)

            # Membuat potongan 5 detik
            chunks = make_chunks(audio, chunk_length_ms)  # 5000 ms = 5 detik

            # Menyimpan setiap potongan audio ke folder output
            for i, chunk in enumerate(chunks):
                output_filename = f"{os.path.splitext(filename)[0]}_chunk_{i}.wav"
                output_path = os.path.join(output_folder, output_filename)
                chunk.export(output_path, format="wav")
                print(f"Potongan disimpan di: {output_path}")

# Contoh penggunaan
input_folder = "/content/drive/MyDrive/Dataset Burung Resampling/Cacomantis Merulinus"  # Ganti dengan path folder input Anda
output_folder = "/content/drive/MyDrive/Dataset Burung Segmentasi/Cacomantis Merulinus"  # Ganti dengan path folder output Anda
split_audio(input_folder, output_folder)
```

### Cacomantis Merulinus

```python
import librosa
import librosa.display
import os
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat Mel-spektrogram
def create_mel_spectrogram(audio_path, output_path, sr=16000, n_fft=1024, hop_length=512, n_mels=128):
    try:
        # Load file audio
        signal, sr = librosa.load(audio_path, sr=sr)

        # Buat Mel-spektrogram
        mel_spectrogram = librosa.feature.melspectrogram(
            y=signal, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels
        )

        # Konversi ke dB
        mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

        # Plot Mel-spektrogram
        # Plot Mel-spektrogram
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(
            mel_spectrogram_db,
            sr=sr,
            hop_length=hop_length,
            x_axis='time',
            y_axis='mel',
            cmap='magma'
        )
        plt.axis('off')  # Menghapus axis dan elemen lainnya
        plt.tight_layout()

        # Simpan gambar
        plt.savefig(output_path)
        plt.close()
        print(f"Mel-spektrogram berhasil disimpan: {output_path}")
    except Exception as e:
        print(f"Error memproses {audio_path}: {e}")

# Fungsi untuk memproses semua file dalam folder
def process_folder(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)  # Buat folder output jika belum ada
    for filename in os.listdir(input_folder):
        if filename.endswith('.wav'):  # Proses hanya file WAV
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.png")
            create_mel_spectrogram(input_path, output_path)

# Contoh penggunaan
input_folder = "/content/drive/MyDrive/Dataset Burung Segmentasi/Cacomantis Merulinus"  # Path folder audio
output_folder = "/content/drive/MyDrive/Dataset Burung Mel-Spektrogram/Cacomantis Merulinus"  # Path folder untuk menyimpan gambar
process_folder(input_folder, output_folder)
```

```python
import librosa
import librosa.display
import os
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat Mel-spektrogram
def create_mel_spectrogram(audio_path, output_path, sr=16000, n_fft=1024, hop_length=512, n_mels=128):
    try:
        # Load file audio
        signal, sr = librosa.load(audio_path, sr=sr)

        # Buat Mel-spektrogram
        mel_spectrogram = librosa.feature.melspectrogram(
            y=signal, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels
        )

        # Konversi ke dB
        mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

        # Plot Mel-spektrogram
        # Plot Mel-spektrogram
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(
            mel_spectrogram_db,
            sr=sr,
            hop_length=hop_length,
            x_axis='time',
            y_axis='mel',
            cmap='magma'
        )
        plt.axis('off')  # Menghapus axis dan elemen lainnya
        plt.tight_layout()

        # Simpan gambar
        plt.savefig(output_path)
        plt.close()
        print(f"Mel-spektrogram berhasil disimpan: {output_path}")
    except Exception as e:
        print(f"Error memproses {audio_path}: {e}")

# Fungsi untuk memproses semua file dalam folder
def process_folder(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)  # Buat folder output jika belum ada
    for filename in os.listdir(input_folder):
        if filename.endswith('.wav'):  # Proses hanya file WAV
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.png")
            create_mel_spectrogram(input_path, output_path)

# Contoh penggunaan
input_folder = "/content/drive/MyDrive/Dataset Burung Segmentasi/Surniculus Lugubris"  # Path folder audio
output_folder = "/content/drive/MyDrive/Dataset Burung Mel-Spektrogram/Surniculus Lugubri"  # Path folder untuk menyimpan gambar
process_folder(input_folder, output_folder)
```

## Subsampling dari Mels Spektrogram

```python
import os
import shutil
import random

# Path asal tempat file berada
source_path = "/content/drive/MyDrive/Dataset Burung Mel-Spektrogram/Surniculus Lugubri"

# Path tujuan untuk menyimpan file yang diacak
output_path = "/content/drive/MyDrive/Dataset Burung Mel-Spektrogram Filter/Surniculus Lugubris"

# Periksa apakah path sumber ada
if not os.path.exists(source_path):
    raise FileNotFoundError(f"Source path '{source_path}' does not exist.")

# Buat folder output jika belum ada
os.makedirs(output_path, exist_ok=True)

# Ambil daftar semua file dalam folder
all_files = [f for f in os.listdir(source_path) if os.path.isfile(os.path.join(source_path, f))]

# Periksa apakah ada file untuk diproses
if not all_files:
    raise FileNotFoundError(f"No files found in source path '{source_path}'.")

# Acak urutan file
random.shuffle(all_files)

# Tentukan jumlah maksimum file yang akan disalin
max_files = 150

# Ambil file pertama sebanyak max_files
selected_files = all_files[:max_files]

# Salin file ke folder tujuan
for file_name in selected_files:
    source_file = os.path.join(source_path, file_name)
    destination_file = os.path.join(output_path, file_name)
    shutil.copy(source_file, destination_file)

print(f"Proses selesai. {len(selected_files)} file telah disalin ke folder '{output_path}'.")
```
